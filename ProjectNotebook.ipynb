#Finish Preprocessing Data

```python
#re-read the data
breast_cancer_dataset_final = pd.read_csv("https://raw.githubusercontent.com/IsitaT03/HSEP_Fall_2022/main/Breast_Cancer.csv")
#tumor size
#Grade
#A stage
#diffetnetiation
#race

breast_cancer_for_analysis = breast_cancer_dataset_final[['Tumor Size', 'Grade', 'A Stage', 'differentiate', 'Race']]
```

```python
breast_cancer_for_analysis['Grade'].unique()
```

```python
#preprocess Grade
breast_cancer_for_analysis.loc[breast_cancer_for_analysis["Grade"]==" anaplastic; Grade IV","Grade"]='4'
breast_cancer_for_analysis['Grade'] = pd.to_numeric(breast_cancer_for_analysis['Grade'])

#preprocess A-Stage
breast_cancer_for_analysis.loc[breast_cancer_for_analysis["A Stage"]=="Regional","A Stage"]=1
breast_cancer_for_analysis.loc[breast_cancer_for_analysis["A Stage"]=="Distant","A Stage"]=0

#preprocess diffrentiate
breast_cancer_for_analysis['differentiate'].replace(['Undifferentiated', 'Poorly differentiated', 'Moderately differentiated', 'Well differentiated'], [0, 1, 2, 3], inplace=True)

#preprocess Race
# Get one hot encoding of column "Race"
one_hot = pd.get_dummies(breast_cancer_for_analysis["Race"])
#lose column "Race" as it is  encoded
breast_cancer_for_analysis = breast_cancer_for_analysis.drop("Race",axis = 1)
# Put the new encoded columns back in
breast_cancer_for_analysis = breast_cancer_for_analysis.join(one_hot)

#Output
breast_cancer_dataset_final.loc[breast_cancer_dataset_final["Status"]=="Dead","Status"]=0
breast_cancer_dataset_final.loc[breast_cancer_dataset_final["Status"]=="Alive","Status"]=1

breast_cancer_for_analysis = breast_cancer_for_analysis.join(breast_cancer_dataset_final["Status"])
breast_cancer_for_analysis
```

#Algorithm Time!




```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
from keras.utils import np_utils


# Break the dataset up into the features (X) and their labels (y)
X = breast_cancer_for_analysis.iloc[:, 0:7].values
y = breast_cancer_for_analysis.iloc[:, 7].values
X=normalize(X,axis=0)

# Split up the X and y datasets randomly into train and test sets
# 20% of the dataset will be used for the test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=31)

X_train=np.asarray(X_train).astype(np.float)
y_train=np.asarray(y_train).astype(np.float)
X_test=np.asarray(X_test).astype(np.float)
y_test=np.asarray(y_test).astype(np.float)
```

```python
#ANN

import keras
from keras.models import Sequential 
from keras.layers import Dense,Activation,Dropout 


# Initialising the ANN
model = Sequential()

# Adding the input layer and the first hidden layer
model.add(Dense(1000,input_dim=7,activation='relu'))
model.add(Dense(50,activation='relu'))

#Protects against overfitting
model.add(Dropout(0.2))

# Adding the output layer
model.add(Dense(1,activation='softmax'))

# Compiling the ANN
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=10,epochs=10,verbose=1)
```

```python
# Predicting the Test set results
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

prediction=model.predict(X_test)
length=len(prediction)
y_label=np.argmax(y_test)
predict_label=np.argmax(prediction,axis=1)
#how times it matched/ how many test cases
accuracy=np.sum(y_label==predict_label)/length * 100 
print("Accuracy of the dataset",accuracy )
```
